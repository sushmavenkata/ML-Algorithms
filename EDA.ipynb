```python
'''
Exploratory Data Analysis (EDA) is an approach to analyzing datasets to summarize their main characteristics,
often employing statistical and visualization methods.
It's a crucial initial step in the data analysis processthat helps analysts and data scientists understand the data and
its underlying structure before applying more complex statistical models or machine learning algorithms.
EDA aims to uncover patterns, relationships, anomalies, and insights within the data.
Overall, EDA is a critical phase in understanding the nature and characteristics of the data, guiding subsequent steps in analysis, and
ultimately ensuring that the conclusions drawn from the data are reliable and valid.
It's a flexible and creative process that involves a combination of statistical methods, visualizations, and
domain knowledge to extract meaningful information from the data.
'''
```

```python
'''
In exploratory data analysis (EDA), data can be categorized into various types based on the nature of the information they represent.
The two primary categories are:

Numerical Data:
    Numerical data consists of quantitative information represented by numbers. This category can further be divided into two subtypes:

Continuous Data: These are measurements that can take on any value within a certain range.
    Discrete Data: These are distinct and separate values that usually represent counts or whole numbers.
Categorical Data:
    Categorical data consists of qualitative information that represents categories or groups.

Nominal Data: This represents categories without any specific order or ranking. Examples include gender, colors, types of cars, etc.
Ordinal Data: These categories have a specific order or ranking.
    Examples include education levels (like high school, bachelor's, master's, etc.) or ratings (such as "low," "medium," "high").
'''
```

```python
'''
If there is lots of categories in the data, you need to consider the process of
'recategorization'
'''
def mapping_education(x):
    if x in ['5th-6th','10th','1st-4th','Preschool','12th','11th','9th','7th-8th']:
        return 'low_level_grade'
    elif x in ['HS-grad','Some-college','Assoc-acdm','Assoc-voc']:
        return 'medium_level_grade'
    elif x in ['Bachelors','Prof-school','Masters','Doctorate']:
        return 'high_level_grade'
```

```python
df['education_summary'] = df['education'].apply(mapping_education)
df
```

```python
sns.countplot(x='education_summary',data=df)
```

```python
'''
Positive Correlation (Close to 1): If the correlation coefficient is close to 1, it suggests a positive relationship.
In this context, it would mean that as a person's age increases, there's a tendency for the number of hours they work per week to increase as well.

Negative Correlation (Close to -1): If the correlation coefficient is close to -1, it implies a negative relationship.
In this case, it would mean that as individuals get older, they might tend to work fewer hours per week.

No Correlation (Close to 0): A correlation coefficient near 0 indicates little to no linear relationship between the 'age' and 'hours_per_week' variables.
It suggests that changes in age do not significantly predict changes in the number of hours worked per week, or vice versa.
'''
correlation = df.age.corr(df.hours_per_week)
```
